### Summary
I decided to generate my own data, as I could not find any dataset interesting for this task and I wanted to develop a system that can help me rate movies better. The idea is to decide how to rate a movie on IMDB.

I define following universe variables:
1. Movie length: Can be from 12 up to 300min
2. Genre that I like: From 0 to 5
3. Actors that I like: From 0 to 5
4. Director that I like: From 0 to 5
5. Story that I like: From 0 to 5

Each of them has their own membership: Low, Medium, High

I choose these values because normally we evaluate aspects of a movie on a very discrete manner. "I like that actor", "I am indifferent towards that director", etc

For the knowledge base, see the comment section below. The test and the values generated can be seen in the carts below too.


### Conclusion
In conclusion,  fuzzy networks is a good way to represent decision systems. However, it seems this can only work for very simplistic and well-predefined decisions. In this example, I rated with 6.5 plot and director. The network suggested I should give the movie a rating of 4 out of 10.

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
'''
Generate universe variables
* Genre, actors, director
and plot on subjective ranges [0, 5]
* Rating has a range of [1, 10] in units of points
* Length has a range from [20, 300] in minutes
'''

x_genre = np.arange(0, 11, 1)
x_actor = np.arange(0,11, 1)
x_director = np.arange(0, 11, 1)
x_plot = np.arange(0, 11, 1)
x_length = np.arange(20, 301, 1)
x_rating  = np.arange(0, 11, 1)

# Generate fuzzy membership functions
genre_lo = fuzz.trimf(x_genre, [0, 0, 5])
genre_md = fuzz.trimf(x_genre, [0, 5, 7])
genre_hi = fuzz.trimf(x_genre, [8, 10, 10])

actor_lo = fuzz.trimf(x_actor, [0, 0, 3])
actor_md = fuzz.trimf(x_actor, [0, 6, 8])
actor_hi = fuzz.trimf(x_actor, [7, 10, 10])

director_lo = fuzz.trimf(x_director, [0, 0, 3])
director_md = fuzz.trimf(x_director, [0, 6, 8])
director_hi = fuzz.trimf(x_director, [7, 10, 10])

plot_lo = fuzz.trimf(x_plot, [0, 0, 4])
plot_md = fuzz.trimf(x_plot, [6, 7, 8])
plot_hi = fuzz.trimf(x_plot, [9, 10, 10])

length_lo = fuzz.trimf(x_length, [20, 40, 300])
length_md = fuzz.trimf(x_length, [85, 90, 100])
length_hi = fuzz.trimf(x_length, [100, 120, 140])

rating_lo = fuzz.trimf(x_rating, [0, 1, 2])
rating_md = fuzz.trimf(x_rating, [3, 4, 5])
rating_hi = fuzz.trimf(x_rating, [6, 8, 10])

# Visualize these universes and membership functions with some plots
fig, (ax0, ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=6, figsize=(8, 9))

ax0.plot(x_qual, genre_lo, 'y', linewidth=1.5, label='Bad')
ax0.plot(x_qual, genre_md, 'b', linewidth=1.5, label='Decent')
ax0.plot(x_qual, genre_hi, 'g', linewidth=1.5, label='Great')
ax0.set_title('How good is the genre?')
ax0.legend()

ax1.plot(x_qual, actor_lo, 'y', linewidth=1.5, label='Bad')
ax1.plot(x_qual, actor_md, 'b', linewidth=1.5, label='Decent')
ax1.plot(x_qual, actor_hi, 'g', linewidth=1.5, label='Great')
ax1.set_title('How good is the actor?')
ax1.legend()

ax2.plot(x_qual, director_lo, 'y', linewidth=1.5, label='Bad')
ax2.plot(x_qual, director_md, 'b', linewidth=1.5, label='Decent')
ax2.plot(x_qual, director_hi, 'g', linewidth=1.5, label='Great')
ax2.set_title('How good is the director?')
ax2.legend()

ax3.plot(x_qual, plot_lo, 'y', linewidth=1.5, label='Bad')
ax3.plot(x_qual, plot_md, 'b', linewidth=1.5, label='Decent')
ax3.plot(x_qual, plot_hi, 'g', linewidth=1.5, label='Great')
ax3.set_title('How good is the plot?')
ax3.legend()

ax4.plot(x_serv, serv_lo, 'b', linewidth=1.5, label='Poor')
ax4.plot(x_serv, serv_md, 'g', linewidth=1.5, label='Acceptable')
ax4.plot(x_serv, serv_hi, 'r', linewidth=1.5, label='Amazing')
ax4.set_title('Movie length')
ax4.legend()

ax5.plot(x_rating, rating_lo, 'b', linewidth=1.5, label='Low')
ax5.plot(x_rating, rating_md, 'g', linewidth=1.5, label='Medium')
ax5.plot(x_rating, rating_hi, 'r', linewidth=1.5, label='High')
ax5.set_title('Rating level')
ax5.legend()

# Do some style adjustments
for ax in (ax0, ax1, ax2, ax3, ax4, ax5):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
# We ativation of our fuzzy membership functions at these values.
# The exact values 6.5 and 6.5 do not exist on our universes...
# This is what fuzz.interp_membership exists for!
plot_level_lo = fuzz.interp_membership(x_plot, plot_lo, 6.5)
plot_level_md = fuzz.interp_membership(x_plot, plot_md, 6.5)
plot_level_hi = fuzz.interp_membership(x_plot, plot_hi, 6.5)

director_level_lo = fuzz.interp_membership(x_director, director_lo, 6.5)
director_level_md = fuzz.interp_membership(x_director, director_md, 6.5)
director_level_hi = fuzz.interp_membership(x_director, director_hi, 6.6)

# Now we take our rules and apply them. Rule 1 concerns bad plot OR bad director.
# With OR operator we take the maximum of these two.
active_rule1 = np.fmax(plot_level_lo, director_level_lo)
rating_activation_lo = np.fmin(active_rule1, rating_lo)

# For rule 2 we connect acceptable plot to medium rating
rating_activation_md = np.fmin(plot_level_md, rating_md)

# For rule 3 we connect high plot OR high director with high rating
active_rule3 = np.fmax(plot_level_hi, director_level_hi)
rating_activation_hi = np.fmin(active_rule3, rating_hi)
rating0 = np.zeros_like(x_rating)

# Visualize this
fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.fill_between(x_rating, rating0, rating_activation_lo, facecolor='b', alpha=0.7)
ax0.plot(x_rating, rating_lo, 'b', linewidth=0.5, linestyle='--', )
ax0.fill_between(x_rating, rating0, rating_activation_md, facecolor='g', alpha=0.7)
ax0.plot(x_rating, rating_md, 'g', linewidth=0.5, linestyle='--')
ax0.fill_between(x_rating, rating0, rating_activation_hi, facecolor='r', alpha=0.7)
ax0.plot(x_rating, rating_hi, 'r', linewidth=0.5, linestyle='--')
ax0.set_title('Output membership activity')

# Turn off top/right axes
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
# Aggregate all three output membership functions together
aggregated = np.fmax(rating_activation_lo,
                     np.fmax(rating_activation_md, rating_activation_hi))

# Calculate defuzzified result
rating = fuzz.defuzz(x_rating, aggregated, 'centroid')
rating_activation = fuzz.interp_membership(x_rating, aggregated, rating)  # for plot

# Visualize this
fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.plot(x_rating, rating_lo, 'b', linewidth=0.5, linestyle='--', )
ax0.plot(x_rating, rating_md, 'g', linewidth=0.5, linestyle='--')
ax0.plot(x_rating, rating_hi, 'r', linewidth=0.5, linestyle='--')
ax0.fill_between(x_rating, rating0, aggregated, facecolor='Orange', alpha=0.7)
ax0.plot([rating, rating], [0, rating_activation], 'k', linewidth=1.5, alpha=0.9)
ax0.set_title('Aggregated membership and result (line)')

# Turn off top/right axes
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```
